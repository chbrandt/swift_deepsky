# Handling `browse`'s Swift observations table to `SDS` master table

```python
import os
assert os.path.isfile('4carlos.txt')
```

```python
with open('4carlos.txt','r') as f:
    head = f.readline()
    top1 = f.readline()
    top2 = f.readline()
print(head)
print(top1)
print(top2)
```

```python
names = [n.strip() for n in head.split('<|>')]
names
```

```python
print('Number of column names:',len(names))
```

```python
print('Number of column fields:',len([c.strip() for c in top2.split('<|>')]))
```

```python
import pandas as pd

df = pd.read_csv('4carlos.txt', delimiter='<\|>', skiprows=2, names=names)

df['START_TIME'] = pd.to_datetime(df['START_TIME'], dayfirst=True)
df['STOP_TIME'] = pd.to_datetime(df['STOP_TIME'], dayfirst=True)
df['ARCHIVE_DATE'] = pd.to_datetime(df['ARCHIVE_DATE'].str.replace('\s+',''), dayfirst=True, errors='coerce')
df['ORIG_OBSID'] = df['ORIG_OBSID'].apply(lambda s:'{:011d}'.format(s))
df['OBSID'] = df['OBSID'].apply(lambda s:'{:011d}'.format(s))

df_obj = df.select_dtypes(['object'])
df[df_obj.columns] = df_obj.apply(lambda x:x.str.replace('\s+',''))
```

```python
df
```

```python
df.to_csv('SwiftXrt_master.csv', sep=';', index=None, date_format='%d/%m/%Y', float_format='%.5f')
```
